```python
vocali = ["a", "e", "i", "o", "u"]
consonanti = ["b", "c", "d", "f", "g", "l", "m", "n", "p", "r", "s", "t", "v", "z"]


testa_monoconsonantica = ["CV", "CVC"]
testa_biconsonantica = ["CCV", "CSV", "CSVC", "CCVC"]
testa_triconsonantica = []
coda_monoconsonantica = ["CSVC", "CCVC", "VC", "CVC"]
coda_biconsonantica = []

combinazioni = {
    "testa_biconsonantica": 
                 ["pr", "pl", "ps", "pi", "pu",                               # 1
                  "br", "bl", "bs", "bi", "bu",                               # 1
                  "fr", "fl", "fs", "fi", "fu",                               # 1
                  "vr", "vl", "vs", "vi", "vu",                               # 1
                  "zi",                                                       # 2
                  "sb", "sd", "sf", "sl", "sm", "sn", "sp",                   # 3
                  "sr", "st", "sv", "sc", "sg",                               # 3
                  "mi", "mu", "ni", "nu", "li", "lu", "ri", "ru"],            # 4
    "coda_monoconsonantica": ["n", "l", "s", "r"],
    "geminazione": consonanti,
}

sc_sg = ["sca", "sco", "scu", "sche", "schi",                        # 3
         "sga", "sgo", "sgu", "sghe", "sghi"]
```

```python
def genera_lista(numero_parole, numero_sillabe, ultima_lettera, soglia_geminazione, debug=False):
    
    global percentuali, vocali, consonanti, testa_monoconsonantica, testa_biconsonantica, \
           coda_monoconsonantica, combinazioni, sc_sg
    
    
    for _ in range(numero_parole):
        sillabe = [0]*numero_sillabe
        geminazione = [0]*(numero_sillabe-1)

        dizionario_scelte = percentuali[numero_sillabe]

        for pos, _ in enumerate(sillabe):

            items = dizionario_scelte[pos].items()
            population = [x for x,y in items]
            distribution = [y for x,y in items]
            distribution = [x/sum(distribution) for x in distribution]

            sillaba_scelta = random.choices(population, distribution)[0]

            sillabe[pos] = sillaba_scelta


        for pos in range(len(sillabe)-1):
            if sillabe[pos] in coda_monoconsonantica and sillabe[pos+1] in testa_monoconsonantica:
                if random.random() > soglia_geminazione:
                    geminazione[pos] = 1

        traduzione = []

        consonante_geminata = None
        for n, sillaba in enumerate(sillabe):

            if sillaba in testa_monoconsonantica:
                if consonante_geminata:
                    testa = consonante_geminata
                else:
                    testa = random.choice(consonanti)
            elif sillaba in testa_biconsonantica:
                testa = random.choice(combinazioni["testa_biconsonantica"])
                if testa in ["sc", "sg"]:
                    vocali = ["a", "o", "u"]
            elif sillaba in testa_triconsonantica:
                pass
            else: # NON HA TESTA
                testa = ''

            consonante_geminata = None

            nucleo = random.choice(vocali)

            if sillaba in coda_monoconsonantica:
                if geminazione[n]:
                    coda = random.choice(consonanti)
                    consonante_geminata = coda
                else:
                    coda = random.choice(combinazioni["coda_monoconsonantica"])
            elif sillaba in coda_biconsonantica:
                pass
            else: # NON HA CODA
                coda = ''

            traduzione.append(testa+nucleo+coda)
            vocali = ["a", "e", "i", "o", "u"]

        if traduzione[-1][-1] in vocali:
            traduzione[-1] = traduzione[-1][:-1]+_ULTIMA_LETTERA

        if debug:
            print(sillabe, geminazione, "".join(traduzione))
        else:
            print("".join(traduzione))
```

```python
random.seed(42)

_NUMERO_SILLABE = 4
_NUMERO_PAROLE = 200
_ULTIMA_LETTERA = "a"
_SOGLIA_GEMINAZIONE = 0.5


genera_lista(_NUMERO_PAROLE, _NUMERO_SILLABE, _ULTIMA_LETTERA, _SOGLIA_GEMINAZIONE, True)
```
